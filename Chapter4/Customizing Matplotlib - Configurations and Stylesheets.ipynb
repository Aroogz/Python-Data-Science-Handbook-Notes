### Plot Customization by Hand

```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np

%matplotlib inline
```

```python
x = np.random.randn(1000)
plt.hist(x);
```

__let's do some Modification__

```python
ax = plt.axes(facecolor='#E6E6E6')
ax.set_axisbelow(True)

# draw solid white grid lines
plt.grid(color='w', linestyle='solid')

# hide axis spines
for spine in ax.spines.values():
    spine.set_visible(False)
    
# hide top and right ticks
#ax.xaxis.tick_bottom()
#ax.yaxis.tick_left()
ax.tick_params(left='on', right='off', top='off')

# lighten ticks and labels
ax.tick_params(colors='gray', direction='out')
for tick in ax.get_xticklabels():
    tick.set_color('gray')
for tick in ax.get_yticklabels():
    tick.set_color('gray')

# control face and edge color of histogram
ax.hist(x, edgecolor='#E6E6E6', color='#EE6666');
```

### Changing the Defaults: rcParams (runtime configurations)

```python
# first save params for editing
IPython_default = plt.rcParams.copy() 
```

```python
# now changing with plt.rc
from matplotlib import cycler
colors = cycler('color', 
               ['#EE6666', '#338888', '#9988DD',
               '#EECC55', '#88BB44', '#FFBBBB'])
plt.rc('axes',facecolor='#E6E6E6', edgecolor='none',
      axisbelow=True, grid=True, prop_cycle=colors)

plt.rc('axes', facecolor='#E6E6E6', edgecolor='none',
axisbelow=True, grid=True, prop_cycle=colors)
plt.rc('grid', color='w', linestyle='solid')
plt.rc('xtick', direction='out', color='gray')
plt.rc('ytick', direction='out', color='gray')
plt.rc('patch', edgecolor='#E6E6E6')
plt.rc('lines', linewidth=2)
```

```python
plt.hist(x);
```

```python
# let's see line plots
for i in range(4):
    plt.plot(np.random.rand(10))
```

### Stylesheets
Provide usefule stylesheets

```python
plt.style.available[:7]
```

Setting a style using plt.style.use('stylename') changes the style for the rest of the session.
For a temporary setting:
    `with plt.style.context('stylename')`

#### Default style

```python
def hist_and_lines():
    np.random.seed(0)
    fig, ax = plt.subplots(1,2, figsize=(11,4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a','b','c'], loc='lower left')
```

```python
# reset rcparams
plt.rcParams.update(IPython_default);
```

```python
hist_and_lines()
```

__FiveThirtyEight style__

```python
with plt.style.context('fivethirtyeight'):
    hist_and_lines()
```

__ggplot__

```python
with plt.style.context('ggplot'):
    hist_and_lines()
```

__Bayesian Methods for Hackers style__

```python
with plt.style.context('bmh'):
    hist_and_lines()
```

__Dark Background__ (for presentation)

```python
with plt.style.context('dark_background'):
    hist_and_lines()
```

__Grayscale__

```python
with plt.style.context('grayscale'):
    hist_and_lines()
```
