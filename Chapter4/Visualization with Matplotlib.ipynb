#### Show() or No show()? how to display plots

##### Plotting from a script

```python
# from a script
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

__Note One thing to be aware of: the plt.show() command should be used only once per Python session, and is most often seen at the very end of the script. Multiple show() commands can lead to unpredictable backend-dependent behavior, and should mostly be avoided__

#### Plotting from an IPython notebook

  - %matplotlib inline will lead to static images of your plot embedded in the

```python
# generally we will usually need inline
# it needs to be done only once per kernel/session
%matplotlib inline
x = np.linspace(0,10,100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

### Two Interfaces for the price of One

#### MATLAB-style interface

MATLAB-style tools are contained in the
pyplot (plt) interface.

```python
plt.figure() # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2,1,1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create second panel plot and ser current axis
plt.subplot(2,1,2)
plt.plot(x, np.cos(x));
```

THis method is stateful as it keeps track of the "current" figure and axes which are where all plt commands are applied.
Reference can be made with `plt.gcf` and `plt.gca` routines

However there is problem accessing a figure once another is created

#### Object-oriented interface
This provides better and fine-grained control

```python
# first createa a grid of plots
# ax will be an array of two axes objects
fig, ax = plt.subplots(2)

# call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
