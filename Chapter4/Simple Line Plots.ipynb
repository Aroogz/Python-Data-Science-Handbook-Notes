```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

__we usually start by creating a figure and axes__

```python
fig = plt.figure()
ax = plt.axes()
```

__we can then use `ax.plot`__

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x));
ax.plot(x, np.cos(x))
```

__alternatively using the pylab interface__

```python
plt.plot(x, np.sin(x)); # convinience like qplot in ggplot
```

```python
# multiple plot using the pylab interface
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

### Adjusting the Plot: Line Colors and Styles

```python
# e.g
plt.plt(x, np.sin(x - 0), color='blue')    # color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
# line style keyword
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
# or combining line and color
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

#### Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

__if reverse is needed for some reason (lol)__

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

`plt.axis` (note this is with the 'i'), makes it even easire

```python
plt.plot(x, np.sin(x))

plt.axis([-1, 11, -1.5, 1.5]) # specify limit as list
```

```python
# automatically tighten the bounds

plt.plot(x, np.sin(x))
plt.axis('tight'); 
```

```python
# aspect ratio equal
plt.plot(x, np.sin(x))
plt.axis('equal');
```

#### Labelling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A sine curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

__legend by specifying label in plot command__

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') # specifying label
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

For transitioning between MATLAB-style functions and object-oriented methods,
make the following changes:
- plt.xlabel() → ax.set_xlabel()
- plt.ylabel() → ax.set_ylabel()
- plt.xlim() → ax.set_xlim()
- plt.ylim() → ax.set_ylim()
- plt.title() → ax.set_title()

__using the OOP plotting however:__

```python
ax = plt.axes()
ax.plot(x, np.sin(x))

# it's easier to just use the set method instead of calling each func as in other
ax.set(xlim=(0,10), ylim=(-2,2),
      xlabel='x', ylabel='sin(x)',
      title='A Simple Plot');
```
